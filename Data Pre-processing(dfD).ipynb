```python
col = ['age','watched_percentage']
```

```python
for c in col:
    per25 = dfD[c].quantile(0.25)
    per75 = dfD[c].quantile(0.75)
    iqr = per75 - per25
    upper_lt = per75 + (0.5*iqr)
    lower_lt = per25 - (0.5*iqr)
    dfD = dfD[dfD[c]<upper_lt]
    dfD = dfD[dfD[c]>lower_lt]
    plt.figure
    sns.boxplot(y=c, data=dfD)
```
